<h2 id="Transform">Transform</h2>
<a href="#Sumary">Sumary</a>

<br/><p>Here, all transformations are carried out on the data so that it is better analyzed and does not bring possible errors due to inconsistencies in the data.</p>

The function below was developed because the pandas replace did not bring the expected result, the deletion of characters.

```python
def removing_substring(for_list):
    list_changed = []
    for i in for_list:
        final_string = ''
        for j in i:
            if j == "'":
                continue
            elif j == "[":
                continue
            elif j == "]":
                continue
            else:
                final_string += j
        list_changed.append(final_string)
    return list_changed
```

```python
list_genres_changed = removing_substring(genres_list)
```

```python
len(list_genres_changed)
```

```python
list_production_countries_changed = removing_substring(production_countries_list)
```

```python
len(list_production_countries_changed)
```

```python
df_genres_changes = pd.DataFrame(list_genres_changed)
df_production_countries = pd.DataFrame(list_production_countries_changed)
```

```python
df_concat = pd.concat([df_genres_changes, df_production_countries], axis = 1, ignore_index = True)
```

```python
df_concat.rename({0:'genres', 1:'production_countries'}, axis='columns', inplace=True)
```

```python
df_copy_titles.drop('genres', axis = 1, inplace = True)
```

```python
df_copy_titles.drop('production_countries', axis = 1, inplace = True)
```

```python
df_copy_titles = pd.concat([df_copy_titles, df_concat], axis = 1)
```

```python
df_copy_titles.replace('', 'null', inplace = True)
```

```python
df_shows = df_copy_titles[df_copy_titles["type"].str.contains("SHOW") == True]
```

```python
df_movies = df_copy_titles[df_copy_titles["type"].str.contains("MOVIE") == True]
```
